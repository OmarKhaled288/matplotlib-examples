```python
t = np.arange(0, 2.5, 0.1)
y1 = np.sin(np.pi*t)
y2 = np.sin(np.pi*t + np.pi/2)
y3 = np.sin(np.pi*t - np.pi/2)
plt.plot(t, y1, 'r*', t,y2,'g^', t,y3,'bx')
plt.legend(['first', 'second', 'third'],loc=1)
```

```python
dx = 0.01
dy = 0.01
x = np.arange(-2.0, 2.0, dx)
y = np.arange(-2.0, 2.0, dy)
x,y = np.meshgrid(x,y)
def f(x,y):
    return(1-y**5 + x**5)*np.exp(-x**2-y**2)
plt.contourf(x,y,f(x,y),8)
```

```python
fig = plt.figure()
ax = Axes3D(fig)
dx = 0.01
dy = 0.01
x = np.arange(-2.0, 2.0, dx)
y = np.arange(-2.0, 2.0, dy)
x,y = np.meshgrid(x,y)
def f(x,y):
    return(1-y**5 + x**5)*np.exp(-x**2-y**2)
ax.plot_surface(x,y,f(x,y), cstride = 1, rstride = 1)
ax.show()
```
